```python
def organizeDF(df):
    test = df[['Manufacturer','no.labs','1_Mean%HBA1c','1_Mean bias','1_%cv']]
    test2 = df[['Manufacturer','no.labs','2_Mean%HBA1c','2_Mean bias','2_%cv']]
    test3 = df[['Manufacturer','no.labs','3_Mean%HBA1c','3_Mean bias','3_%cv']]
    test4 = df[['Manufacturer','no.labs','4_Mean%HBA1c','4_Mean bias','4_%cv']]
    test5 = df[['Manufacturer','no.labs','5_Mean%HBA1c','5_Mean bias','5_%cv']]
    test2 = test2.rename(columns= {'2_Mean%HBA1c':'1_Mean%HBA1c','2_Mean bias':'1_Mean bias','2_%cv':'1_%cv'})
    test3 = test3.rename(columns= {'3_Mean%HBA1c':'1_Mean%HBA1c','3_Mean bias':'1_Mean bias','3_%cv':'1_%cv'})
    test4 = test4.rename(columns= {'4_Mean%HBA1c':'1_Mean%HBA1c','4_Mean bias':'1_Mean bias','4_%cv':'1_%cv'})
    test5 = test5.rename(columns= {'5_Mean%HBA1c':'1_Mean%HBA1c','5_Mean bias':'1_Mean bias','5_%cv':'1_%cv'})
    test6 = pd.concat([test,test2])
    test7 = pd.concat([test6,test3])
    test8 = pd.concat([test7,test4])
    test9 = pd.concat([test8,test5])
    test10 = test9.rename(columns= {'1_Mean%HBA1c':'Mean%HBA1c','1_Mean bias':'Mean bias','1_%cv':'%cv'})
    return test9
```
