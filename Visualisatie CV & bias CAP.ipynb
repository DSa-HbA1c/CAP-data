```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import plotly.express as px
import cufflinks as cf
import numpy as np
cf.go_offline()
cf.set_config_file(offline=False, world_readable=True)
```

```python
df = pd.read_excel('CAP-samenvoeging.xlsx')
df
```

```python
df = df.dropna()
df.info()
```

```python
# df.astype({'CV': 'float64'})
# df.astype({'Bias': 'np.float64'}).dtypes
# df['CV'] = df['CV'].astype(np.float64)
df['Bias'] = df['Bias'].astype(float)
```

```python
df.info()
```

```python
df2 = df.groupby(['Method name','Year']).mean().reset_index()
df3 = df2.set_index('Method name')
df2
```

```python
def plotCV(method):
    df3.loc[[method]].iplot(
                    xTitle='Year',
                    x = 'Year',
                    y = 'CV',
                    mode='lines+markers',
                    yTitle='CV {}'.format(method)
                    
        
                     )
```

```python
interact(plotCV, method=df3.index.unique())
```
